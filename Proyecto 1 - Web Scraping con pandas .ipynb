# 2 Leer .csv de multiples URLs con pandas 

- https://www.football-data.co.uk/mmz4281/2425/E0.csv
- https://www.football-data.co.uk/mmz4281/2425/E1.csv

```python
# Estructura de un link 
"https://www.football-data.co.uk/mmz4281/" + "2425" + "/" + "E0" + ".csv"
```

```python
# crear una variable root 
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiples ligas (Links)

- https://www.football-data.co.uk/mmz4281/2425/E0.csv
- https://www.football-data.co.uk/mmz4281/2425/E1.csv
- https://www.football-data.co.uk/mmz4281/2425/E2.csv
- https://www.football-data.co.uk/mmz4281/2425/E3.csv
- https://www.football-data.co.uk/mmz4281/2425/EC.csv

```python
#  Crear Listas de ligas 
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# Hacer un loop en la lista ligas, leer multiples csv y adjuntarlo a una lista 
for league in leagues: 
    df = pd.read_csv(root + '2425' + '/' + league + '.csv')
    frames.append(df)
```

```python
# tamaño de los frames 
len(frames)
```

```python
# Mostras 1er, 2do y 3er elementos
frames[4]
```

## 2.2 Multiples Temporadas

```python
#  Crear Listas de ligas 
leagues = ['E0', 'E2', 'E3']
frames = []

# Hacer loop a un rango de temporadas e insertar una nueva columna 
for league in leagues: 
    for season in range(19, 25):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# Total de frames (6 temporadas x 3 ligas)
len(frames)
```

```python
# Mostrar 1er y ultimo elemento (columna season fue creada )
frames[17]
```

# 3 Organizar Data en un Diccionario 

```python
# Crear un diccionario con nombre original de liga como llave 
dict_countries = {'Spanish La Liga': 'SP1', 'Spanish Segunda Division':'SP2', 'German BundesLiga' : 'D1', 'English Premier League': 'E0', 'English Premier League 1': 'E2', 'English Premier League 2': 'E3'}
```

```python
# obtener un elemento del diccionario
dict_countries['Spanish La Liga']
```

```python
dict_historical_data={}

#  Crear Listas de ligas 
#leagues = ['E0', 'E2', 'E3']

# hacer loop sobre llaves de diccionario 
for league in dict_countries: 
    frames = []
    for season in range(19, 25):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# mostrar dataframe dentro del dist_historical_data
dict_historical_data['English Premier League']
```
